```python
import tensorflow as tf

# Carrega dataset MNIST
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# Normaliza imagens: escala os valores dos pixels de 0-255 para 0-1
train_images = train_images / 255.0
test_images = test_images / 255.0

modelo = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

modelo.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Treina o modelo
modelo.fit(train_images, train_labels, epochs=10, verbose=1)

# Avalia o modelo
loss, accuracy = modelo.evaluate(test_images, test_labels, verbose=1)
print(f"AcurГЎcia no conjunto de teste: {accuracy:.2%}")

# Exemplo de previsГЈo
print("Fazendo uma previsГЈo para a primeira imagem de teste:")
predictions = modelo.predict(test_images[:1])
predicted_label = tf.argmax(predictions[0]).numpy()
print(f"Label verdadeiro: {test_labels[0]}")
print(f"Label previsto: {predicted_label}")
```
