```python
import pandas as pd
from groq import Groq

# Criando conexão a API
client = Groq(
    api_key=""
)

# DataFrame
df = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos"],
    "vendas": [200, 350, 150]
})

# Transformar tabela em texto
dados = df.to_string(index=False)

# Pergunta
pergunta = "Quem vendeu mais?"

# Chamada para IA
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
            "role": "system",
            "content": "Você analisa tabelas do pandas."
        },
        {
            "role": "user",
            "content": f"""

            Tabela:
            {dados}

            Pergunta:
            {pergunta}

            """
        }
    ]
)

# Resultado
print(response.choices[0].message.content)
```
