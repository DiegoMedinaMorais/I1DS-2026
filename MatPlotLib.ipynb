```python
# matplotlib
import numpy as np
import matplotlib.pyplot as plt

# 1. Dados de exemplo (Simulação)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1) # Equação alvo: y = 4 + 3x

# 2. Parâmetros da "IA" (Ex: Regressão Linear Simples)
# Suponha que o modelo aprendeu os pesos y_pred = 3.5 * X + 3.8
X_novo = np.array([[0], [2]])
y_pred = 3.5 * X_novo + 3.8

# 3. Visualização
plt.figure(figsize=(8, 6))

# Plotando os dados reais
plt.scatter(X, y, color='blue', alpha=0.6, label='Dados Reais (Treinamento)')

# Plotando a linha prevista pelo modelo
plt.plot(X_novo, y_pred, color='red', linewidth=2, label='Predição da IA')

# 4. Customização do Gráfico
plt.title('Avaliação de Modelo: Dados vs. Predição')
plt.xlabel('Variável de Entrada (X)')
plt.ylabel('Variável Alvo (y)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Exibir o gráfico
plt.show()
```
